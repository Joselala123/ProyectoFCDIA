# Lectura de csv


```python
import pandas as pd
```

```python
df=pd.read_csv("EDV_2022.csv",sep=";")
```

# Limpieza de datos

Utilizando **convert_dtypes()** se transforma las columnas de tipo Object de ser posible a String o int

```python
df=df.convert_dtypes()
```

Las columnas de tipo String se estandarizan con **strip() y casefold()** para eliminar posibles errores de tipeo e inconsistencias en el texto.

```python
for serie in df.columns:
  if df[serie].dtype=="String":
    df[serie]=df[serie].str.strip().str.casefold()
```

```python
df.isna().sum() #Niguna fila contiene datos nulos.
```

```python
print(df.duplicated().sum()) #Puesto que la suma da cero esto quiere decir que cada dato es unico.
```

Datos atipicos

```python
listaColumnasNumericas=df.select_dtypes(include='int64').columns.tolist() #Se selecciona las columnas numericas del df
print(listaColumnasNumericas)
listaColumnasNumericas.remove("anio_insc")
listaColumnasNumericas.remove("dia_insc")

for columna in listaColumnasNumericas:
  print("COLUMNA: "+columna)
  serie=df[columna]
  q1=serie.quantile(0.25)
  q3=serie.quantile(0.75)
  iqr=q3-q1
  limInferior=q1-1.5*iqr
  limSuperior=q3+1.5*iqr
  datosAtipicos=serie[(serie<limInferior) | (serie>limSuperior)]
  print(datosAtipicos.head())
  print(datosAtipicos.tail())
```
